```python
from keras.models import Sequential
from keras.layers import Conv2D
import numpy as np
import pandas as pd
import cv2
from sklearn.model_selection import train_test_split
import os
```

```python
def load_images_from_folder(lowres):
    images = []
    for root, dirs, files in os.walk(lowres):
        for filename in files:
            if filename.endswith(".jpg"):
                img_path = os.path.join(root, filename)
                img = cv2.imread(img_path)
                if img is not None:
                    images.append(img)
    return images
```

```python
pics_folder=r"APP\1d\lowres"
images=load_images_from_folder(pics_folder)
```
